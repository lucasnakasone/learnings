```python
import random
from IPython.display import clear_output

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
# Controls overall game loop
game_on = True

class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit
        self.value = values[rank]
    
    def __str__(self):
        return f"{self.rank} of {self.suit}"
        
class Deck:
    def __init__(self):
        #print("Deck created...")
        #print("Cards created...")
        self.cards = []
        for rank in ranks:
            for suit in suits:
                self.cards.append(Card(rank, suit))
        
    def shuffle(self):
        random.shuffle(self.cards)
        
    def deal(self):
        return self.cards.pop() 
                
class Hand:
    def __init__(self):
        #print("Hand created...\n")
        self.cards = []
        self.value = 0            
        self.aces = 0
        
    def add_card(self, card):
        self.cards.append(card)
        self.value += card.value
        if card.rank == "Ace":
            self.aces += 1
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1 
    
    def show_cards(self):
        [print(f"{each}") for each in self.cards]
        
class Player:
    def __init__(self):
        #print("Player created...")
        self.pot = Pot(100)
        self.hand = Hand()

class Dealer:
    def __init__(self):
        #print("Dealer created...")
        #self.pot = Pot(10000)
        self.deck = Deck()
        self.hand = Hand()
                
    def take_bet(self):
        while True:
            try:
                player.pot.bet = int(input('How many chips would you like to bet? '))
            except ValueError:
                print('Sorry, a bet must be an integer!')
            else:
                if player.pot.bet > player.pot.chips:
                    print("Sorry, your bet can't exceed",player.pot.chips,"chips.")
                else:
                    break
    
    def hit(self, player):
        player.hand.add_card(dealer.deck.deal())
        player.hand.adjust_for_ace()
    
    def hit_or_stand(self, hand):
        global game_on  
        while True:
            x = input("\nWould you like to (H)it or (S)tand?: ")
            if x[0].lower() == 'h':
                dealer.hit(hand)
            elif x[0].lower() == 's':
                print("Player stands. Dealer is playing...")
                game_on = False
            else:
                print("Sorry, please try again.")
                continue
            break
    
    # The functions below sets the end of the round
    def player_busts(self, player):
        print("Player busts!")
        print(f"You reached {player.hand.value} points!")
        player.pot.lose_bet()

    def player_wins(self, player):
        print("Player wins!")
        player.pot.win_bet()

    def dealer_busts(self, player):
        print("\nDealer busts!")
        print(f"Dealer reached {player.hand.value} points!")
        player.pot.win_bet()
    
    def dealer_wins(self, player):
        print("Dealer wins!")
        player.pot.lose_bet()
    
    def push(self):
        print("Dealer and Player tied! It's a push!")
        print(f"Both reached {dealer.hand.value} points!")
                    
class Pot:
    def __init__(self, chips):
        self.chips = chips
        self.bet = 0
        #print("Pot created...")
        
    def win_bet(self):
        self.chips += self.bet
    
    def lose_bet(self):
        self.chips -+ self.bet
```
